```python
from openai import OpenAI
import os
from dotenv import load_dotenv

# load environment variable from .env file
load_dotenv()

# Access variables using os.getenv()

openai_key = os.getenv("OPENAI_API_KEY")

language = "Japaneese"
# Japaneese, Chineese

max_words=50

openai_client = OpenAI(api_key=openai_key)

# babbage-002, gpt-3.5-turbo, gpt-3.5-turbo-0125, gpt-4.1-nano

"""
                How do you say, 'Hello' in {language}? (what that word means - and why there are multiple words (if any).
                Also What is their favoriate dish, according to the cities (if city specific are more famous).
                And What are their most favourite - tourist places + cities (in Bullet points)?
"""
try:
    response = openai_client.chat.completions.create(
        model="gpt-3.5-turbo-0125",
        messages=[
            {
                "role": "user",
                "content": f"""
                Name 5 cool muslim boy names (ONLY GIVE NAMES, No other text)

                Write in {max_words} words
                """
            }
        ],
        max_tokens=max_words,
        # n = 2
    )
    # print("RESPONSE=========", response)
    # print("CHOICES========\n", response.choices)
    # print("MESSAGE========\n", response.choices[0].message)
    print("MESSAGE CONTENT========\n", response.choices[0].message.content)
    # print("MESSAGE ROLE========\n", response.choices[0].message.role)
    # print("MODEL========\n", response.model)
    print("OUTPUT TOKENS========\n", response.usage.completion_tokens)
    print("INPUT TOKENS========\n", response.usage.prompt_tokens)
    # print("CHAT ID========\n", response.id)





except Exception as e:
    print(f"Error: {e}")
```
